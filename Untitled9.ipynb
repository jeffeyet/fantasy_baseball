```python
import json
import os
from requests_oauthlib import OAuth1Session

# 1) Paste your Consumer Key/Secret here (AFTER you regenerate the secret)
CONSUMER_KEY = os.environ["YAHOO_CONSUMER_KEY"].strip()
CONSUMER_SECRET = os.environ["YAHOO_CONSUMER_SECRET"].strip()

# Yahoo OAuth 1.0a endpoints used in the Fantasy Sports guide
REQUEST_TOKEN_URL = "https://api.login.yahoo.com/oauth/v2/get_request_token"
ACCESS_TOKEN_URL  = "https://api.login.yahoo.com/oauth/v2/get_token"

TOKEN_FILE = "yahoo_tokens.json"

def save_tokens(tokens: dict) -> None:
    with open(TOKEN_FILE, "w") as f:
        json.dump(tokens, f, indent=2)

def load_tokens():
    if os.path.exists(TOKEN_FILE):
        with open(TOKEN_FILE, "r") as f:
            return json.load(f)
    return None

def get_oauth_session(tokens=None):
    if tokens:
        return OAuth1Session(
            CONSUMER_KEY,
            client_secret=CONSUMER_SECRET,
            resource_owner_key=tokens["oauth_token"],
            resource_owner_secret=tokens["oauth_token_secret"],
        )
    return OAuth1Session(CONSUMER_KEY, client_secret=CONSUMER_SECRET, callback_uri="oob")

def main():
    # If we already have tokens, try using them first
    tokens = load_tokens()
    if tokens:
        print("Found saved tokens. We'll try an API call with them.\n")
        sess = get_oauth_session(tokens)
    else:
        # 2) Get request token (callback_uri is set in the session constructor
        #    so it's included in the OAuth signature)
        sess = get_oauth_session()
        rt = sess.fetch_request_token(REQUEST_TOKEN_URL)
        request_token = rt["oauth_token"]
        request_secret = rt["oauth_token_secret"]

        # 3) Send user to authorize URL (Yahoo returns this URL in the response)
        auth_url = rt.get("xoauth_request_auth_url")
        if not auth_url:
            # fallback (rare), but keep you moving:
            auth_url = f"https://api.login.yahoo.com/oauth/v2/request_auth?oauth_token={request_token}"

        print("1) Open this URL in your browser and approve access:\n")
        print(auth_url, "\n")
        verifier = input("2) Paste the verifier code Yahoo shows you here: ").strip()

        # 4) Exchange request token + verifier for access token
        sess = OAuth1Session(
            CONSUMER_KEY,
            client_secret=CONSUMER_SECRET,
            resource_owner_key=request_token,
            resource_owner_secret=request_secret,
            verifier=verifier,
        )
        at = sess.fetch_access_token(ACCESS_TOKEN_URL)
        tokens = {
            "oauth_token": at["oauth_token"],
            "oauth_token_secret": at["oauth_token_secret"],
            # Yahoo may also return oauth_session_handle for token refresh workflows
            "oauth_session_handle": at.get("oauth_session_handle"),
        }
        save_tokens(tokens)
        print("\nSaved access tokens to yahoo_tokens.json\n")

    # 5) First "proof it works" call: list your MLB leagues
    # Yahoo Fantasy Sports API returns XML by default; format=json makes life easier.
    url = "https://fantasysports.yahooapis.com/fantasy/v2/users;use_login=1/games;game_keys=mlb/leagues?format=json"
    resp = sess.get(url)
    print("Status:", resp.status_code)
    print(resp.text[:2000])  # print first chunk so output isn't overwhelming

if __name__ == "__main__":
    main()
```
